```python
from bs4 import BeautifulSoup

with open('schools.html', 'r') as file:
    html = file.read()

#print(html)    
soup = BeautifulSoup(str(html))
links = soup.findAll('a')

data = []
for link in links:
    print(link.text, link['href'])
    data.append({'text':link.text.replace('\n', ' '), 'link':link['href']})
```

```python
import pandas as pd

df = pd.DataFrame(data)
```

```python
# test = 'http://www.usask.ca'
# test2 = test.split('//')[1].split('/')[0]
# test2

f = lambda test: test.split('//')[1].split('/')[0].split(':')[0]



df['root'] = df['link'].apply(f)
#df['root'].unique()
df.to_csv('schools.csv', index=False)
```
